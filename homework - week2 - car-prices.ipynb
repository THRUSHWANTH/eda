````python
def is_valid_date(date_str):
    try:
        pd.to_datetime(date_str)
        return True
    except:
        return False

# Filter out invalid date entries
valid_dates = df['saledate'].apply(is_valid_date)
df_valid_dates = df[valid_dates]

# Convert the valid saledate entries to datetime format
df_valid_dates['saledate'] = pd.to_datetime(df_valid_dates['saledate'])
````

````python
df['mmr'] = pd.to_numeric(df['mmr'], errors='coerce')
non_numeric_mmr = df.loc[pd.to_numeric(df['mmr'], errors='coerce').isna()]['mmr'].unique()
print(non_numeric_mmr)
````

## Question 4. Analyze the relationship between the wholeprice `mmr` and the `salesprice`.
- What is the overall correlation?  
- What happens to the correlation as you increase the maximum selling price window (e.g., 0-10000, 0-20000,...)?
- Create a scatter plot with the relationship.  
- Create a graph that shows how the correlation changes.

````python
overall_correlation = df['mmr'].corr(df['sellingprice'])
print("Overall Correlation between mmr and sellingprice is:", overall_correlation)
````

Sample narrative description.

An overall correlation of 0.9836 between mmr (likely meaning "Manheim Market Report" or some measure of market price) and sellingprice suggests a very strong positive linear relationship between the two variables. This means as one variable increases, the other also tends to increase in a consistent manner.

In practical terms, if mmr represents some kind of market estimation or average price for a vehicle, this high correlation suggests that the sellingprice at which vehicles are sold is very close to this market estimate. The vehicles' selling prices are closely aligned with their market values.

This kind of information can be valuable for both sellers and buyers in the market. Sellers can use the mmr as a reliable guide for setting their selling prices, while buyers can use it to ensure they're getting a fair deal.







````python
import matplotlib.pyplot as plt
import numpy as np

# Define price windows
windows = [10000, 20000, 30000, 40000, 50000]  # ... and so on
correlations = []

# Calculate correlation for each window
for window in windows:
    segment = df[df['sellingprice'] <= window]
    corr = segment['mmr'].corr(segment['sellingprice'])
    correlations.append(corr)

# Scatter plot for a selected window (e.g., 0-20,000)
selected_segment = df[df['sellingprice'] <= 20000]
plt.scatter(selected_segment['mmr'], selected_segment['sellingprice'])
plt.title('Scatter Plot of MMR vs Selling Price (0-20,000)')
plt.xlabel('MMR')
plt.ylabel('Selling Price')
plt.show()

# Line/Bar graph showing how correlation changes with selling price window
plt.plot(windows, correlations, marker='o')
plt.title('Correlation between MMR and Selling Price over Different Price Windows')
plt.xlabel('Price Window')
plt.ylabel('Correlation')
plt.show()
````

the MMR is a strong predictor of selling price across varying price windows, especially as the price range increases. However, there are specific price segments where the correlation might slightly vary, suggesting external factors influencing the selling price. The presence of data points with high MMR but lower selling prices in the scatter plot also emphasizes the need to consider other factors when pricing vehicles.

## Question 5. How does the distribution of `salesprice` vary by make for cars made between 2010-2015?
- Use the 10 most common makes (your cleaned version of this column).  
- Create a graph that answers this question.

````python
# Filter the dataframe for cars made between 2010-2015
filtered_df = df[(df['year'] >= 2010) & (df['year'] <= 2015)]
````

````python
# Get the 10 most common makes
top_makes = filtered_df['make'].value_counts().head(10).index.tolist()

# Filter the dataframe to only include rows with the top 10 makes
top_makes_df = filtered_df[filtered_df['make'].isin(top_makes)]
top_makes_df
````

Sample narrative description.

````python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,10))
sns.boxplot(data=top_makes_df, x='make', y='sellingprice')
plt.title('Distribution of Sales Price by Make (2010-2015)')
plt.xlabel('Make')
plt.ylabel('Sales Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````
